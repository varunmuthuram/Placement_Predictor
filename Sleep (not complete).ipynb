```python
x = data.drop("15. How would you rate your overall academic performance (GPA or grades) in the past semester?", axis =1)
y = data["15. How would you rate your overall academic performance (GPA or grades) in the past semester?"]
```

```python
from sklearn.preprocessing import OneHotEncoder

encoder = OneHotEncoder(drop='first')
x_encoded = encoder.fit_transform(x)

from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
y_encoded = le.fit_transform(y)
```

```python
x_encoded
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x_encoded, y_encoded, test_size=0.2)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
clf = RandomForestClassifier()
```

```python
clf.fit(x_train, y_train)
```

```python
clf.score(x_test, y_test)
```
