```python
import pandas as pd
import numpy as np
```

```python
placement = pd.read_csv("Placement_BeginnerTask01.csv")
```

```python
placement
```

```python
placement.drop("StudentID", axis=1, inplace=True)
```

```python
placement.dtypes
```

```python
placement["ExtracurricularActivities"] = placement["ExtracurricularActivities"].astype("category")
placement["PlacementTraining"] = placement["PlacementTraining"].astype("category")
placement["PlacementStatus"] = placement["PlacementStatus"].astype("category")
```

```python
placement.dtypes
```

```python
x = placement.drop("PlacementStatus", axis=1)
y = placement["PlacementStatus"]
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer

cat_cols = ("ExtracurricularActivities", "PlacementTraining")
transformer = ColumnTransformer([("cols", OneHotEncoder(), cat_cols)], remainder = "passthrough")
transformed_x = transformer.fit_transform(x)

transformed_x
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(transformed_x, y, test_size=0.2)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
clf = RandomForestClassifier()
```

```python
clf.fit(x_train, y_train)
```

```python
y_preds = clf.predict(x_test)
clf.score(x_test, y_test)
```

```python
from sklearn.model_selection import cross_val_score
cross_val_score(clf, transformed_x, y, cv=5)
mean = np.mean(cross_val_score(clf, transformed_x, y, cv=5))
mean
```

```python
from sklearn.linear_model import LogisticRegression

clf1 = LogisticRegression(max_iter=1000)
```

```python
clf1.fit(x_train, y_train)
```

```python
y_preds1 = clf1.predict(x_test)
clf1.score(x_test, y_test)
```

```python
mean1 = np.mean(cross_val_score(clf1, transformed_x, y, cv=5))
mean1
```
